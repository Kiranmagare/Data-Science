### 1. column normalization:(minmax scaling)
- we are bringing the entire feature values into the range of 0 to 1--[0,1]
- feature---->a1,a2,a3-------an
- normlization--->a1',a2',a3',--------an'
- find the max and min element in the column

    * a1'=a1-min(col)/max(col)-min(col)
    * amax=a9=(a9-amin)/(amax-amin)=1
    * amin=a7=(a7-amin)/(amax-amin)=0.
- syntax:from sklearn.preprocessing import MinMaxScaler
- whenever we have a column with a different measurement
- weight in kg,height in cm
- it will all the values of the column in the range of [0,1]
- If we are using the ML which more sensitive towards the measurement or distance--then in that case also we will use column normalization.
- ex:KNN,kmeans etc
- it is better to use the when the data is sparse
- ex:dating apps:interest matrix
* tinder:
* interests:

      name     singing    dancing   cooking   breathing
      mayur     0           0         0         1
      xyz       1           1         1         0
      pqr       0           1         1         0

      mayur     0           0         0         1
- most of the values in the column are 0-->not having the particular info.--sparse data
- if the data is sparse we can use the column normalization.

- dense data:max values in the columns are non zeros
- we have a lots of information.


- ex:pizza:sparse pizza--->sauce+1

       overloded--->dense
- ex:features:weight  height

### encoding:

  - Converting the categorical column into numerical columns
  - different types of categorical data.
   1. ordinal : category is having some order
   2. nominal: we dont have order

##### encoding are of two types :
    
##### 1. label encoding:ordinal data
  
   - we are using label encoding for ordinal data
   - when we have only two categories and data is nominal then we can use label encoding
   - if we have more than two categgories and data is norminal then we can use label encoding
   -  if we have more than two categories we cannot used label encoding for categorical data.
   -  it will convert the categorical column into numerical column
   -  as compare to one hot encoding label encoding is memory effiecient
   
##### 2. onehot encoding:nominal data

   - it is also used to convert categorical data into numerical data.
   - it is mostly used for nominal data
   - it is not memory efficient compare to label encoding becoz it will create a seperate column for each category

       * Syntax: from sklearn.preprocessing import OneHotEncoder
       * en=OneHotEncoder(sparse_output=False)
       * arr=en.fit_transform(df[[]])
