### Gaussian Distribution:

   - gaussian distribution is continuous probability distribution whose PDF is symmetric about means.
   -  X---->continuous random variable RV
   -  gaussian distribution ==normal distribution
   -  X~GD(mean, variance) whose probability density function is bell shaped
   -  distribution parameters---mean, variance

   * ex: Iris data set---
          - PL----gaussian distribution
          - PL(mean,variance)--- PDF

### skewness:
  - how far the distribution is from symmetric

      1. left skewed (-ve skewed)
      2. right skewed(+ve skewed)
      3. symmetric

  - if the distribution is left or right skewed we have to convert it to the symmetric distribution(gaussian/normal)
  - for that we have to do the transformation 

 1. log transformation(right skewed):
     y=log(x)
     x=price
     prediction --->exp(y)

 2. reciporcal transformation (right skewed):
     y=1/x 
    prediction:1/y  

 3. squareroot transformation:(right skewed):
    y=sqrt(x)
    prediction:y^2

 4. box cox transformation(right and left skewed)

### Uniform Distribution
   - equiprobable events or equally likely events
   - if the outcomes are having the equal chances of happening then that distribution generally a uniform distribution 
   - each outcome have uniform or exactly same value
 
  * ex: rolling a die getting probability

        getting(1)=1/6
        getting(2)=1/6
        getting(3)=1/6
        getting(4)=1/6
        getting(5)=1/6
        getting(6)=1/6----equally likely ----uniform distribution

   * ex. rolling a two die and getting a sum----not equally likely event so not following the uniform distribution

##### Q. ex. rolling a die probability of getting a prime number is uniform distribution?

   * it is not a uniform distribution, because ---->if the probability of prime number is renormalized into the subset of prime number such as {2,3,5} then in this case all the three outcomes are Queally likely and they are uniformly distributed but the total sample space of die this events are not a equally likely so rolling a die and getting a prime number is not a uniform distribution  

##### Q. when to count expected value or normal mean in uniform distribution
